```python
textos = pd.read_csv(r'bases\austen_books.csv')
textos
```

## Removendo Nulos

```python
textos_na = textos.dropna()
textos_na
```

## Aplicando Análise de Sentimento Biblioteca VaderSentiment

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
```

```python

def calculate_sentiment_scores(sentence):
    sntmnt = analyzer.polarity_scores(sentence)['compound']
    return(sntmnt)
```

```python
emma_book = textos_na[textos_na['book'] == 'Emma']
```

```python
eng_snt_score =  []

for comment in emma_book.text.to_list():
    snts_score = calculate_sentiment_scores(comment)
    eng_snt_score.append(snts_score)
```

```python
emma_book['sentiment_score'] = np.array(eng_snt_score)
emma_book
```

## Criando Array categorizando o Score de Sentimento

```python
i = 0

vader_sentiment = [ ]

while(i<len(emma_book)):
    if ((emma_book.iloc[i]['sentiment_score'] >= 0.05)):
        vader_sentiment.append('positive')
        i = i+1
    elif ((emma_book.iloc[i]['sentiment_score'] > -0.05) & (emma_book.iloc[i]['sentiment_score'] < 0.05)):
        vader_sentiment.append('neutral')
        i = i+1
    elif ((emma_book.iloc[i]['sentiment_score'] <= -0.05)):
        vader_sentiment.append('negative')
        i = i+1
```

```python
emma_book['sentiment'] = vader_sentiment
emma_book
```
