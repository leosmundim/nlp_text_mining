```python
#from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, WordPunctTokenizer
```

# Text Mining - Português

Referência: https://blog.caelum.com.br/entendendo-unicode-e-os-character-encodings/amp/

```python
dados_totais = pd.read_csv(r'bases\portuguese-poems.csv')
dados_totais
```

### Removendo os Acentos

```python
from unidecode import unidecode

poetas = dados_totais['Author'].apply(lambda w: unidecode(w))
```

```python
poetas
```

## Unnest

```python
all_tokens = []
def tokenize_text(text):
    tokens = word_tokenize(text.lower(), 
                           language='portuguese')
    all_tokens.extend(tokens)
    return tokens

dados_unnested = poetas.apply(tokenize_text) 
```

```python
all_tokens
```

```python
dados_unnested
```

## Remove Stop Words

```python
# Removendo Stop Words

from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('portuguese'))
```

```python
all_tokens = [w for w in all_tokens if not w in stop_words]
```

```python
all_tokens
```

## Autores que mais Aparecem

```python
df_all_tokens = pd.DataFrame(all_tokens, columns=['word'])
df_all_tokens['word'].value_counts().sort_values(ascending=False)
```

## Word Cloud

```python
# Criando Função para Plotar

def plotWordCloud(wc):
    fig, ax = plt.subplots(figsize=(14,7))
    ax.imshow(wc, interpolation = 'bilinear')
    ax.set_axis_off()
    plt.imshow(wc)
```

```python
all_tokens_wc = " ".join(w for w in df_all_tokens['word'])

#Criando máscara
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

wc = WordCloud(background_color='white', 
              width=1600,
              height=800, 
               max_words= 50, 
               max_font_size= 200, 
               mask=mask,
              colormap='viridis').generate(all_tokens_wc)
plotWordCloud(wc)
```
