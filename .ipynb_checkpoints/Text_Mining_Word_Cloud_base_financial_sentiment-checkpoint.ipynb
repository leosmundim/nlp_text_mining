### Unnesting

Sem Pacote

```python
token = "".join(w for w in df).lower().split(" ")
dados_unnested = pd.DataFrame(token, columns=['word'])
dados_unnested
```

Usando Pacote NLTK

```python
all_tokens = []
def tokenize_text(text):
    tokens = word_tokenize(text.lower())
    all_tokens.extend(tokens)
    return tokens

dados_unnested = df.apply(tokenize_text) 
```

```python
all_tokens
```

```python
# Removendo os números da lista de tokens e alguns caracteres não importantes

all_tokens = [w for w in all_tokens if not (w.isdigit() 
                                            or w.replace('.','').isdigit() 
                                            or w == '.' 
                                            or w == '$' 
                                            or w == ',' 
                                            or w == "'s" 
                                            or w == '(' 
                                            or w == ')' 
                                            or w == '%')]
```

```python
all_tokens
```

## Removendo Stop Words

```python
# Removendo Stop Words

from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
```

```python
all_tokens = [w for w in all_tokens if not w in stop_words]
```

```python
all_tokens
```

```python
len(all_tokens)
```

## Analisando as Palavras que mais aparecem

```python
df_all_tokens = pd.DataFrame(all_tokens, columns=['word'])
df_all_tokens['word'].value_counts().sort_values(ascending=False)
```

## Word Cloud

Documentação: https://amueller.github.io/word_cloud/ <br>
Referência: https://sigmoidal.ai/como-criar-uma-wordcloud-em-python/

```python
# Criando Função para Plotar

def plotWordCloud(wc):
    fig, ax = plt.subplots(figsize=(14,7))
    ax.imshow(wc, interpolation = 'bilinear')
    ax.set_axis_off()
    plt.imshow(wc)
```

```python
all_tokens_wc = " ".join(w for w in df_all_tokens['word'])

#Criando máscara
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

wc = WordCloud(background_color='white', 
              width=1600,
              height=800, 
               max_words= 50, 
               max_font_size= 200, 
               mask=mask,
              colormap='viridis').generate(all_tokens_wc)
plotWordCloud(wc)
```

## Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

df_all_tokens_stem = pd.DataFrame([stemmer.stem(w) for w in df_all_tokens['word']], columns=['stem'])
```
