# Text Mining - Correlação entre Palavras

Foi criado um dataframe utilizando o livro da Janet Austin onde tokenizaram as palvras e marcaram em qual sessão cada uma delas aparecia.

```python
df = pd.read_csv(r'bases\austen_section_words.csv')
df
```

```python
from collections import Counter

# Calcule a contagem de pares usando Counter
pair_counts = Counter(zip(df['word']))
```

```python
# Create word pairs
word_pairs = [(df['word'][i], df['word'][i+1]) for i in range(len(df)-1)]

# Count the pairs
word_pair_counts = Counter(word_pairs)
```

```python
word_pair_counts
```

```python
most_common_pairs = word_pair_counts.most_common(10)  # Change 10 to the desired number
print("Most common word pairs:")
for pair, count in most_common_pairs:
    print(pair, count)
```
