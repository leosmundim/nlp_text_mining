```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize, WordPunctTokenizer
```

# NLP - TF-IDF

Referência: https://thinkinfi.com/document-similarity-matching-using-tf-idf-python/

Encontrar as palavras que mais aparece no documento mas que diferencia o documento dos demais.<br>
Objetivo é observar a frequência que a palavra aparece em determinado documento mas com baixa frequência nos demais.<br>


```python
textos = pd.read_csv(r'bases\austen_books.csv')
textos
```

```python
textos_na = textos.dropna()
textos_na
```

```python
# Primeiro, tokenizamos o texto
textos_na['word'] = textos_na['text'].apply(lambda x: word_tokenize(x))

# Em seguida, agrupamos por 'book' e contamos as palavras
word_counts = textos_na.explode('word').groupby(['book', 'word']).size().reset_index(name='count')

# Ordenamos os resultados
word_counts = word_counts.sort_values(by=['book', 'count'], ascending=False)

```

```python
word_counts
```

## TF-IDF

```python
pd.set_option('display.max_columns', 100)
```

```python
textos_na_corpus = textos_na.groupby('book')['text'].agg(lambda x: ' '.join(x)).reset_index()
textos_na_corpus
```

### Removendo Palavras desnecessária que não foram removidas na função TfidfVectorizer(stop_words='english') 

```python
bannedWord = ['Mr', 'Miss', 'Mrs', 'said', 'Sir', 'thing', 'think', 'did']

corpus = textos_na_corpus['text']

for palavra in bannedWord:
    corpus = corpus.str.replace(palavra, "")
    
textos_na_corpus['temp'] = corpus
textos_na_corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = textos_na_corpus['temp']

# Crie um objeto TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Ajuste o modelo aos seus documentos e transforme os documentos em uma matriz TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

```python
# A matriz TF-IDF agora contém os valores TF-IDF para cada palavra nos documentos
# Você pode convertê-lo em um DataFrame do pandas para visualização
import pandas as pd

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_df = tfidf_df.set_index(textos_na_corpus['book'])
tfidf_df
```

```python
sense_tfidf_df = tfidf_df.T['Sense & Sensibility'].sort_values(ascending=True).iloc[-10:]
sense_tfidf_df
```

### Gráfico com as 10 palavras que mais importam no termo de TF-IDF para explicar o livro 

```python
plt.style.use('seaborn')
plt.barh(sense_tfidf_df.index, sense_tfidf_df)
plt.title('Sense & Sensibility')
```

### Gráfico com as 10 palavras mais importantes de cada Livro de acordo com o TF-IDF

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 15))
for book in tfidf_df.T.columns:
    book_graph_tf_idf = tfidf_df.T[book].sort_values(ascending=True).iloc[-10:]
    
    plt.barh(book_graph_tf_idf.index, book_graph_tf_idf, label=book)
    plt.title(book)
    
plt.xlabel('Valor tf_idf')
plt.ylabel('Palavra')
plt.title('Valor tf_idf por Palavra em Cada Livro')
plt.legend()
```

# Criando DataFrame unificado como forma de aprendizagem

```python
book_graph_list = []
book_list = []
for column in tfidf_df.T.columns:
    book_graph_list.append(tfidf_df.T[column].sort_values(ascending=True)[-10:])
    book_list.extend([column] * 10)
```

```python
book_graph_list = pd.DataFrame(book_graph_list).reset_index().drop('index', axis=1)
```

```python
book_graph = book_graph_list.T.stack()
book_graph = book_graph.reset_index().drop('level_1', axis=1)
book_graph.columns = ['word', 'tf_idf']
#book_graph = book_graph.drop_duplicates('word')
book_graph['book'] = book_list
book_graph
```

```python
# Criar um gráfico de barras horizontais para cada livro
fig, ax = plt.subplots(figsize=(10, 15))

for book, group in book_graph.groupby('book'):
    ax.barh(group['word'], group['tf_idf'], label=book)

ax.set_xlabel('Valor tf_idf')
ax.set_ylabel('Palavra')
ax.set_title('Valor tf_idf por Palavra em Cada Livro')
ax.legend()
plt.show()
```

```python
# Obter uma lista de livros únicos
books = book_graph['book'].unique()

# Criar subplots para cada livro
fig, axs = plt.subplots(len(books), figsize=(10, 6*len(books)))

for i, book in enumerate(books):
    ax = axs[i]
    book_data = book_graph[book_graph['book'] == book]
    grouped = book_data.groupby('word')['tf_idf'].sum().reset_index()
    
    ax.barh(grouped['word'], grouped['tf_idf'])
    ax.set_xlabel('Valor tf_idf')
    ax.set_ylabel('Palavra')
    ax.set_title(f'Valor tf_idf por Palavra em {book}')

plt.tight_layout()
plt.show()
```
